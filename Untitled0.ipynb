```python
import pandas as pd

df = pd.read_csv("/content/exercises.csv")

body_part = input("Hangi vücut bölgesi için hareket önerilsin? ")

hareketler = df[df['bodyPart'].str.lower() == body_part.lower()]

if hareketler.empty:
    print(f"{body_part} için hareket bulunamadı.")
else:
    # 5 rastgele hareket seçelim
    secilenler = hareketler.sample(5)
    print(f"{body_part} için önerilen hareketler:")
    for i, row in secilenler.iterrows():
        print(f"- {row['name']} ({row['equipment']})")
```
